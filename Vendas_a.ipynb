```python
import pandas as pd
import seaborn as sns
import missingno as msno
import numpy as np
from matplotlib import pyplot as plt

pd.options.display.max_rows = 1000
pd.options.display.max_columns = 1000

sns.set()  
```

```python
filepath = 'C:/Users/gmsar/Downloads/Vendas_a.xlsx'

df = pd.read_excel(filepath, sheet_name='Sheet1')
```

```python
df.head()
```

```python
df.columns = ["id","vendas","treina","graduacao"]
```

```python
from scipy import stats

r = stats.pearsonr(df['treina'], df['vendas'])[0] 
text = f'r = {r.round(2)} e r-quadrado = {(r**2).round(2)}'

text
```

```python
ax=sns.regplot(data=df, x="treina", y="vendas", ci=None, line_kws={"color": "red", "linestyle" : "dashed"})
ax.set_ylabel('Vendas')
ax.set_xlabel('Treinamento em anos')


```

```python
ax = sns.boxplot(y='treina', x='graduacao', data=df, orient='v')
ax.set_title('Relação entre tempo no estágio e Relacionamento com o supervisor')
ax.set_ylabel('Tempo de estágio em meses')
ax.set_xlabel('Relacionamento com o supervisor')
```
